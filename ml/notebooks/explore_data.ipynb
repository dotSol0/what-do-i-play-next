```python
import inspect
import imslp # Replace 'my_package' with the name of the package you want to inspect

functions_list = inspect.getmembers(imslp, inspect.isfunction)

# 'functions_list' will be a list of tuples, where the first element is the 
# function name (string) and the second is the function itself (object).

for name, function in functions_list:
    print(name)
```
